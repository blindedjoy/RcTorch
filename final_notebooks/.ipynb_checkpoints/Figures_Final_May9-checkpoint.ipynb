```python
def prediction_plot(b, b_outIC, pathSave=None, exact=False, gts=None):
#     plt.figure(figsize = (6,5))    
    time = b["time"] ;     preds = b["ypreds"] ;       
    y_outIC = b_outIC["outBO_ypreds"]["ypreds"]; 
    gts_outIC = b_outIC["outBO_ypreds"]["gts"]; 
    t_out = b_outIC["time"]

    gts = b["gts"];
    
    alp=0.8
    plt.plot(time, preds[0], linewidth = lineW+1, label = "RC (with BO)", color = "blue")
    plt.plot(t_out, y_outIC[0], linewidth = lineW+2, color = "green", label = "RC (without BO)")
    if exact:
        plt.plot(time, gts[0], '--',linewidth = lineW, label = "ground truth", color = "red", alpha=alp)
        plt.plot(t_out, gts_outIC[0], '--',linewidth = lineW,  color = "red", alpha=alp)
    plt.legend(frameon=False)
    
    for i in range(1, len(gts)):
        plt.plot(time, preds[i], linewidth = lineW+2, color = "blue")
        if exact:
            plt.plot(time, gts[i],'--',linewidth = lineW, color = "red", alpha=alp)

    for i in range(1, len(gts_outIC)):
        plt.plot(t_out, y_outIC[i], linewidth = lineW+2, color = "green",alpha=alp)
        if exact:
            plt.plot(t_out, gts_outIC[i], '--',linewidth = lineW,  color = "red", alpha=1)
            
    ax = plt.gca()
    for ps in ['top','bottom','left','right']:
        ax.spines[ps].set_linewidth(lineBoxW)

    plt.ylabel(r'$y(t)$'); # plt.xlabel(r'$t$');
    plt.tight_layout()
    if not pathSave==None:
        plt.savefig(pathSave,format='eps')
        
        
# Stands for: Root Mean Square Residuals
def RMSR_plot(b, x2=False, forceCoef=1):
#     plt.figure(figsize = (6,3))
    time = b["time"]; y = b["ypreds"];
    gts = b["gts"]; ydot=b["ydots"]    
    y_out = b["outBO_ypreds"]["ypreds"]; ydot_out = b["outBO_ypreds"]["ydots"]    
    
    coef = torch.ones_like(ydot[0])
        
    if x2:
        coef = time.pow(2)

    R, Rout = 0, 0
    
    for i in range(1, len(gts)):
        resids = (ydot[i] + coef*y[i] - forceCoef*torch.sin(time))**2
        resids_out = (ydot_out[i] + coef*y_out[i] - forceCoef*torch.sin(time))**2

        R += resids
        Rout += resids_out

    R = R/len(gts);    R = torch.sqrt(R)    
    plt.plot(time, R, linewidth = lineW, color='blue')
    
    Rout = Rout/len(gts);    Rout = torch.sqrt(Rout)
    plt.plot(time, Rout, linewidth = lineW, color='green')
    plt.yscale("log")
    
    ax = plt.gca()
    for ps in ['top','bottom','left','right']:
        ax.spines[ps].set_linewidth(lineBoxW)

    plt.xlabel(r'$t$'); plt.ylabel(r'{RMSR}'); plt.tight_layout()
```

```python
def prediction_bernoulli(b,b_lin, pathSave=None, exact=False, clr=None):
#     plt.figure(figsize = (6,5))
    time = b["time"];  preds = b["ys"];  gts = b["gts"];  preds_lin = b_lin["linear_trajectories"]

    if not clr:
        clr = ['blue', 'blue', 'blue','blue']
    
    for i in range(0, len(gts)):
        plt.plot(time, preds[i], linewidth = lineW+2, color = clr[i])
        plt.plot(time, preds_lin[i],'--', linewidth = lineW+2, color = clr[i])
                                  
    ax = plt.gca()
    for ps in ['top','bottom','left','right']:
        ax.spines[ps].set_linewidth(lineBoxW)

#     plt.xlabel(r'$t$'); 
    plt.ylabel(r'$y(t)$');   plt.tight_layout()
    
    if not pathSave==None:
        plt.savefig(pathSave,format='pdf')

        
def loss_plots(b, pathSave=None, exact=False, clr=None):
#     plt.figure(figsize=[5,5])

    loss_args = b["loss"]
    
    for i, loss_dict in enumerate(loss_args):
        lr = loss_dict["lr"]
        epochs = loss_dict["epochs"]
        loss_history = loss_dict["loss_history"]
        plt.loglog(loss_history,  color = clr[i], linewidth = lineW)              

    ax = plt.gca()
    for ps in ['top','bottom','left','right']:
        ax.spines[ps].set_linewidth(lineBoxW)

    plt.xlabel(r'Iterations'); plt.ylabel(r'{Loss}');  plt.tight_layout()
    
    if not pathSave==None:
        plt.savefig(pathSave,format='pdf')

# Stands for: Root Mean Square Residuals
def RMSR_plot_bernoulli(b, clr=None):
#     plt.figure(figsize = (6,3))
    time = b["time"]; y = b["ys"]; gts = b["gts"]; ydot=b["ydots"]    

    q=0.5 ;  R = 0
    
    for i in range(0, len(gts)):
        resids = torch.abs( ydot[i] + y[i] + q*y[i]**2 )
        plt.plot(time, resids,   linewidth = lineW, color = clr[i])
        R += resids
#     R = R/len(gts);    R = torch.sqrt(R)    
#     plt.plot(time, R, linewidth = lineW, color='blue')
    plt.yscale("log")
    ax = plt.gca()
    for ps in ['top','bottom','left','right']:
        ax.spines[ps].set_linewidth(lineBoxW)

    plt.xlabel(r'$t$'); plt.ylabel(r'{Residuals}'); plt.tight_layout()
```

```python
# fig = plt.figure(figsize = (8,8))   
fig = plt.figure(figsize = (12,6))   
clr = ['blue', 'green', 'orange','red']

gs1 = gridspec.GridSpec(3, 4)
# gs1.update(left=0.05, right=0.48, wspace=0.05)

ax=plt.subplot(gs1[:-1, :-2])
prediction_bernoulli(b, b_lin, clr=clr)
# ax = plt.xticks([], [])
plt.xticks([0, 1,2, 3,4, 5  ])
ax.tick_params(labelbottom=False)    
# plt.axis('tight')

plt.subplot(gs1[-1, :-2])
RMSR_plot_bernoulli(b, clr=clr)
# plt.axis('tight')
plt.ylim(0,0.05)
plt.xticks([0, 1,2, 3,4, 5  ])

plt.subplot(gs1[0:, -2:])
loss_plots(b, clr=clr)#, pathSave='bernoulli_loss.eps')

plt.xticks([1,  10**2, 10**4])

pathSave='bernoulli_fig.pdf';
# plt.savefig(pathSave,format='pdf');
```

```python
    
def plot_bo_trajectory(dic, BO_index=12):
    time = dic["time"]; x = dic["xs"];  p = dic["ps"]; 
    plt.plot(time/np.pi, x[BO_index],'-b', linewidth=lineW, label = 'x(t)')
#     plt.plot(time/np.pi, p[BO_index],'--b', linewidth=lineW, label = 'p(t)')
    plt.axvline(6, linewidth=lineW, color='r', linestyle='--' )
#     plt.legend(bbox_to_anchor=(1., 1., 1.2,1.2), frameon=False)
#     plt.legend(frameon=False, ncol=2, bbox_to_anchor=(0, 1,1.002, 1.002),)
    ax = plt.gca()
    
    for ps in ['top','bottom','left','right']:
        ax.spines[ps].set_linewidth(lineBoxW)

#     plt.xlabel(r'$t$')
    plt.ylabel(r'$x$'); 
    plt.tight_layout()
    
def res_NLosc(dic, BO_index=12):
    time = dic["time"]; x = dic["xs"];  p = dic["ps"]; 
    xdot = dic["xdots"];  pdot = dic["pdots"]; 
    res_1 = torch.abs( xdot[BO_index] - p[BO_index])
    res_2 = torch.abs(pdot[BO_index] + x[BO_index] + x[BO_index].pow(3))
    res_tot = res_1 + res_2
    plt.plot(time/np.pi, res_tot,'-b', linewidth=lineW)
    plt.axvline(6, linewidth=lineW, color='r', linestyle='--' )
    plt.yscale("log")

    ax = plt.gca()
    for ps in ['top','bottom','left','right']:
        ax.spines[ps].set_linewidth(lineBoxW)
    plt.xlabel(r'$t/\pi$'); plt.ylabel(r'{Residuals}'); plt.tight_layout()

def phaseSpace_NLosc(dic, totIC=10, BO_index=12):
    time = dic["time"]; x = dic["xs"];  p = dic["ps"]; 
    for i in range(0, totIC, 1):
        plt.plot( x[i],p[i],'-g', linewidth=lineW, label = 'x(t)')
        if i==BO_index:
            plt.plot( x[i],p[i],'-b', linewidth=lineW+4, label = 'x(t)')
        ax = plt.gca()
    for ps in ['top','bottom','left','right']:
        ax.spines[ps].set_linewidth(lineBoxW)
    plt.xlabel(r'$x(t)$'); plt.ylabel(r'$p(t)$'); plt.tight_layout()
    plt.axis('equal')
```

```python
BO_index = 6


fig = plt.figure(figsize = (12,6))   
# clr = ['blue', 'green', 'orange','red']

gs1 = gridspec.GridSpec(2,4)


ax=plt.subplot(gs1[:-1, :-2])
plot_bo_trajectory(b, BO_index)
plt.xticks([0, 2, 4, 6, 8, 10  ])
ax.tick_params(labelbottom=False)    


plt.subplot(gs1[-1, :-2])
res_NLosc(b, BO_index); plt.ylim(0,0.02)
plt.xticks([0, 2, 4, 6, 8, 10  ])

plt.subplot(gs1[0:, -2:])
phaseSpace_NLosc(b, totIC=10, BO_index=BO_index)
plt.xticks([-2, 0, 2])
plt.yticks([-2, 0, 2])
plt.tight_layout()
pathSave='nlOsc_fig.pdf';
# plt.savefig(pathSave,format='pdf');
```
